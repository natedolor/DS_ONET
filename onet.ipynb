```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.onetonline.org/find/all'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
soup
```

```python
soup.find('table', class_ = "table tablesorter tablesorter-bootstrap table-responsive-md w-100 table-fluid list-results")
```

```python
occupation_table_title = [title.text for title in soup.find_all('th')]
                          #soup.find_all('th')[0].text
occupation_table_title
```

```python
import pandas as pd

df1 = pd.DataFrame(columns = occupation_table_title)
df2 = pd.DataFrame(columns = ['Job Zone Link', 'Code Link', 'Occupation Link'])

column_data = soup.find_all('tr')
column_data 

for row in column_data:
    row_data = row.find_all('td')
    individual_row = [data.text.strip() for data in row_data]
    if (len(individual_row) == 0):
        continue

    length = len(df1)
    df1.loc[length] = individual_row

for row in column_data:
    individual_link_row = [a['href'] for a in row.find_all('a')]
    if (len(individual_link_row) == 0):
        continue
    if (len(individual_link_row) == 1):
        individual_link_row.append('Not available')
        individual_link_row.insert(0, 'Not available')
    if (len(individual_link_row) == 2):
        if ('skill' in individual_link_row[0] ):
            individual_link_row.append('Not available')
        else:
           individual_link_row.insert(0, 'Not available') 
    
    length = len(df2)
    df2.loc[length] = individual_link_row

df1
df2
```

```python

occupations = pd.concat([df1, df2], axis = 1)
display(occupations)
```
